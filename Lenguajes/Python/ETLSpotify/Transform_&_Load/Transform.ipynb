```python
import pandas as pd
pathData = 'C:\\Users\\dz09d\\Documents\\DataEng\\Proyectos-Practicas-Data-Engineer\\Lenguajes\\Python\\ETLSpotify\\Extract\\archivo_salida.csv'
dataset = pd.read_csv(pathData)
dataset
```

```python
# Contar la cantidad de géneros únicos
cantidad_generos = dataset['genre'].nunique()

generos_existentes= dataset['genre'].unique()

print(f'La cantidad de generos que hay son: {cantidad_generos}')

for genero in generos_existentes:
    print('\n'+ genero)
```

```python
# Contar la cantidad de artistas únicos
cantidad_artistas = dataset['artists_name'].nunique()

print(f"Cantidad de artistas únicos: {cantidad_artistas}")
```

```python
# Agrupar por 'artists_name' y contar las canciones únicas
artistas_con_multiple_canciones = dataset.groupby('artists_name')['name'].nunique()

# Filtrar artistas con más de una canción
artistas_con_mas_de_una_cancion = artistas_con_multiple_canciones[artistas_con_multiple_canciones > 1]

# Obtener los nombres de los artistas con más de una canción
nombres_artistas_multiples_canciones = artistas_con_mas_de_una_cancion.index.tolist()

# Contar cuántos artistas tienen más de una canción
cantidad_artistas_multiples_canciones = artistas_con_mas_de_una_cancion.count()

# Mostrar los resultados
print(f"Cantidad de artistas con más de una canción: {cantidad_artistas_multiples_canciones}")
print("\nNombres de los artistas con más de una canción:")
for artista in nombres_artistas_multiples_canciones:
    print(artista)
```

```python

# Asegurarse de que la columna 'album_release_date' esté en formato de fecha
dataset['album_release_date'] = pd.to_datetime(dataset['album_release_date'], errors='coerce')

# Encontrar el índice de la canción más antigua (con la fecha más temprana)
indice_mas_viejo = dataset['album_release_date'].idxmin()

# Obtener la canción más vieja, el artista y la popularidad
cancion_mas_vieja = dataset.loc[indice_mas_viejo, 'name']
artista_mas_viejo = dataset.loc[indice_mas_viejo, 'artists_name']
popularidad_mas_vieja = dataset.loc[indice_mas_viejo, 'popularity']
fecha_mas_vieja = dataset.loc[indice_mas_viejo, 'album_release_date']

# Mostrar los resultados
print(f"El tema más viejo es '{cancion_mas_vieja}' de {artista_mas_viejo}.")
print(f"Fecha de lanzamiento: {fecha_mas_vieja.strftime('%Y-%m-%d')}")
print(f"Popularidad: {popularidad_mas_vieja}")
```

```python

# Ordenar el dataset por la columna 'popularity' de manera descendente
top_5_popularidad = dataset.sort_values(by='popularity', ascending=False).head(5)

# Seleccionar solo las columnas relevantes: 'name', 'artists_name' y 'popularity'
top_5_popularidad = top_5_popularidad[['name', 'artists_name', 'popularity']]

# Mostrar el top 5
print("Top 5 canciones con más popularidad:")
print(top_5_popularidad)
```
